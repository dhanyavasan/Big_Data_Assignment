```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import requests
response = requests.get('https://www.sfu.ca/computing/people/faculty.html') 
```

```python
#print(response)
table=pd.DataFrame(columns=['Name','Designation','Area']) #table 
```

```python

soup = BeautifulSoup(response.content)


for num,a in enumerate(soup.findAll('div', attrs={'class','textimage section'})):
    t=a.find_all('h4')
    name=t[0].text.split(',')[0]
    Desig=t[0].text.split(',')[1]
    aslot=a.find_all('p')
    area=aslot[0].text.split(':')
    field=None
    if(len(area)==2):
        field=area[1]
    table.loc[num+1]=[name,Desig,field]
```
